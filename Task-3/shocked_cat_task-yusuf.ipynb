```python
import cv2
import matplotlib.pyplot as plt

# Görüntü yolunu belirleyin
image_path = 'Desktop/OPENCV/shocked_cat.jpg'

# Görüntüyü oku ve gri tonlamaya çevir
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Görüntüyü siyah-beyaz olarak göster
plt.imshow(img, cmap='gray')
plt.title('Shocked Cat - Black & White')
plt.axis('off')
plt.show()

# İkili eşikleme uygulama
_, binary_img = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

# Görüntüyü göster
plt.imshow(binary_img, cmap='gray')
plt.title('Binary Thresholded Image')
plt.axis('off')
plt.show()

# Yatay Sobel Filtresi uygulama
sobel_horizontal = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=5)

# Görüntüyü göster
plt.imshow(sobel_horizontal, cmap='gray')
plt.title('Horizontal Sobel Filtered Image')
plt.axis('off')
plt.show()
```
