```python
import cv2
import matplotlib.pyplot as plt
```

```python
monkey_image = cv2.imread('C:/Users/TAS KAFA/Downloads/last_numpy/TASK3/shocked_monkey.jpg')
monkey_rgb = cv2.cvtColor(monkey_image, cv2.COLOR_BGR2RGB)
plt.title('RGB')
plt.imshow(monkey_rgb)
plt.axis('off')
```

```python
shocked_guy = cv2.imread('C:/Users/TAS KAFA/Downloads/last_numpy/TASK3/shocked_guy.png')
shocked_guy_rgb = cv2.cvtColor(shocked_guy, cv2.COLOR_BGR2RGB)
plt.imshow(shocked_guy_rgb)
plt.title('RGB')
plt.axis('off')
```

```python
shocked_guy_ters = cv2.flip(shocked_guy_rgb, 0)
plt.imshow(shocked_guy_ters)
plt.title('Ters')
plt.axis('off')
```

```python
shocked_guy_2 = shocked_guy_ters.copy()
cv2.rectangle(shocked_guy_2, pt1=(500,880), pt2=(700,1050), color=(255,0,0), thickness=6)
plt.imshow(shocked_guy_2)
plt.title('Gz Saptanm')
```

```python
shocked_cat = cv2.imread('C:/Users/TAS KAFA/Downloads/last_numpy/TASK3/shocked_cat.jpg', cv2.IMREAD_GRAYSCALE)
shocked_cat_gray = shocked_cat.copy()
plt.imshow(shocked_cat_gray, cmap='gray')
plt.title('Siyah Beyaz')
plt.axis('off')
```

```python
shocked_cat_gray_binary = shocked_cat_gray.copy()
ret,thresh1 = cv2.threshold(shocked_cat_gray_binary,127,255,cv2.THRESH_BINARY)
plt.imshow(thresh1, cmap='gray')
plt.title('Binary')
plt.axis('off')
```

```python
shocked_cat_sobel = thresh1.copy()
cv2.Sobel(shocked_cat_sobel, cv2.CV_64F, 1, 0, ksize=5)
plt.imshow(shocked_cat_sobel)
plt.title('Binary Sobel')
plt.axis('off')
```
