```python
import cv2
import matplotlib.pyplot as plt

# Görüntüyü yükleme
gorsel_yolu = "Desktop/OPENCV/shocked_monkey.jpg"
gorsel = cv2.imread(gorsel_yolu)

# 1. Görüntüyü RGB formatına çevirme
gorsel_rgb = cv2.cvtColor(gorsel, cv2.COLOR_BGR2RGB)

# RGB görüntüsünü gösterme
plt.imshow(gorsel_rgb)
plt.title("RGB Görüntü")
plt.axis("off")
plt.show()

# 2. Mavi kanalı izole etme (Kırmızı ve Yeşil kanalları sıfırla)
mavi_kanal_gorsel = gorsel.copy()
mavi_kanal_gorsel[:, :, 1] = 0  # Yeşil kanalını sıfırla
mavi_kanal_gorsel[:, :, 2] = 0  # Kırmızı kanalını sıfırla

# İzole edilmiş Mavi kanal görüntüsünü gösterme
plt.imshow(cv2.cvtColor(mavi_kanal_gorsel, cv2.COLOR_BGR2RGB))
plt.title("Mavi Kanal")
plt.axis("off")
plt.show()
```
