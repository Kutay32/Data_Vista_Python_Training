```python
import cv2
import matplotlib.pyplot as plt

# Grnty ykleme
gorsel_yolu = "Desktop/OPENCV/shocked_monkey.jpg"
gorsel = cv2.imread(gorsel_yolu)

# 1. Grnty RGB formatna evirme
gorsel_rgb = cv2.cvtColor(gorsel, cv2.COLOR_BGR2RGB)

# RGB grntsn gsterme
plt.imshow(gorsel_rgb)
plt.title("RGB Grnt")
plt.axis("off")
plt.show()

# 2. Mavi kanal izole etme (Krmz ve Yeil kanallar sfrla)
mavi_kanal_gorsel = gorsel.copy()
mavi_kanal_gorsel[:, :, 1] = 0  # Yeil kanaln sfrla
mavi_kanal_gorsel[:, :, 2] = 0  # Krmz kanaln sfrla

# zole edilmi Mavi kanal grntsn gsterme
plt.imshow(cv2.cvtColor(mavi_kanal_gorsel, cv2.COLOR_BGR2RGB))
plt.title("Mavi Kanal")
plt.axis("off")
plt.show()
```
